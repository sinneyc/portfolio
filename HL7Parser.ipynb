```python
import json
file = open("final.hl7",'r')

first = True
final = ""

for line in file:
    if(line[:3] == "MSH"):
        if(first):
            final += "{"
            first = False
        else:
            final += "}\n{"
    
    field = 0
    com = 0
    sub = 0
    message = ""

    for char in line: 
        if char == "|":
            field += 1
            if message != "":
                final += str(field)+'.'+str(com)+'.'+str(sub)+': '+message+"\n"
                message = "" 
            com = 0
            sub = 0
            
        elif char == "^":
            com += 1
            if message != "":
                final += str(field)+'.'+str(com)+'.'+str(sub)+': '+message+"\n"                
                message = ""
            sub = 0
            
        elif char == "~":
            sub +=1
            if message != "":
                final += str(field)+'.'+str(com)+'.'+str(sub)+': '+message+"\n"
                message = ""
        
        else:
            message += char
            
final += "}"             
print(final)
```
